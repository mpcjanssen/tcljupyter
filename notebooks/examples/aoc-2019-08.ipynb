```tcl
proc svgpixel {x y idprefix border color} {
    return "<rect id=\"$idprefix-$x-$y\" x=\"$x\" y = \"$y\" width=\"1\" height=\"1\" style=\"fill:$color;stroke-width:$border;stroke:rgb(255,255,255)\" />"
}
proc range {start end} {
  set res {}
  for {set i $start} {$i < $end+1} {incr i} {
    lappend res $i
  }
  return $res
}
set width 25
set height 6
set layersize [expr {$width*$height}]
set pat [string repeat . $layersize]
set layers [regexp -all -inline $pat $data];
```

## Visualize with Unicode

```tcl
proc visualize-step-text {} {
  set layer 0


  set bitmap [lrepeat 6 [lrepeat 25 \u2b1c]]
    set id [jupyter::html [join $bitmap <br>]]
      yield
  foreach l [lreverse $::layers] {
    incr layer
    foreach y [range 0 5] {
       foreach x [range 0 24] {
       set idx [expr {$y*25+$x} ]
        set col [string index $l $idx]
        switch $col {
          0 {
              lset bitmap $y $x \u2b1c
          }
          1 {
              lset bitmap $y $x \u2b1b
          }
          2 {
              
          
          }
          
        }
      }
    }
    jupyter::updatehtml $id [join $bitmap <br>]
    yield 1
  }
  yield 0
}
coroutine visualize visualize-step-text
while {[visualize]} {}
```

## Visualize with SVG

```tcl
proc visualize-step {} {
  set layer 0
  set svgid [jupyter::html {<svg width="250" height="60"></svg>}]
  yield
  set svg [lrepeat 150 {}]
  foreach l [lreverse $::layers] {
    incr layer
    foreach y [range 0 5] {
       foreach x [range 0 24] {
       set idx [expr {$y*25+$x} ]
        set col [string index $l $idx]
        switch $col {
          0 {
              lset svg $idx [svgpixel  $x $y vis1  0.01 white]
          }
          1 {
              lset svg $idx [svgpixel  $x $y vis1  0.01 black]
          }
          2 {          
          }          
        }
      }
    }
    jupyter::updatehtml $svgid "<b>layer: </b> $layer<br><svg viewbox=\"-1 -1 26 7\">[join $svg \n]</svg>"
    yield 1
  }
  yield 0
}
```

```tcl
coroutine visualize visualize-step
while {[visualize]} {}
```

## Using JS
You can also directly modify the SVG using batched javascript commands

```tcl
proc visualize-step-2 {} {
  set layer 0
  set svg [list {<svg  viewbox="-1 -1 26 7">}]
  set jsid [jupyter::display text/javascript {}]
  foreach y [range 0 5] {
    foreach x [range 0 24] {
        lappend svg [svgpixel $x $y vis2  0.01 black]
    }
  }
  lappend svg </svg>
  set svg [join $svg \n]
  jupyter::html $svg
  
  yield 
  foreach l [lreverse $::layers] {
    incr layer
    set js  [list {
      function updateColor(id,color){
       
        var myrect = document.getElementById(id);

    myrect.style.opacity = color;
    
    } }]
    foreach y [range 0 5] {
       foreach x [range 0 24] {
       set idx [expr {$y*25+$x} ]
        set col [string index $l $idx]
        switch $col {
          0 {
              lappend js "updateColor('vis2-$x-$y', 0.0);"
          }
          1 {
              lappend js "updateColor('vis2-$x-$y', 1.0);"
          }
          
          
        }
      }
    }
    jupyter::updatedisplay $jsid application/javascript [join $js \n];
    yield 1
  }
  yield 0
};
```

```tcl
coroutine visualize-2 visualize-step-2
while {[visualize-2]} {};
```

## With a canvas

```tcl
proc visualize-step-3 {} {
  set layer 0
  set svg [list {<canvas id="cvs"  width="250" height="60">}]
  
  lappend svg </canvas>
  set svg [join $svg \n]
  jupyter::html $svg
  set js  {

        
     }
  
  set jsid [jupyter::display text/javascript $js]
  yield 
  foreach l [lreverse $::layers] {
    incr layer
    set js  [list {
        var c = document.getElementById("cvs")
        var cvsctx = c.getContext("2d");
        cvsctx.fillStyle = "black";
     }]
    foreach y [range 0 5] {
       foreach x [range 0 24] {
       set idx [expr {$y*25+$x} ]
        set col [string index $l $idx]
        set sx [expr {$x*10}]
        set sy [expr {$y*10}]
        switch $col {
          0 {
             lappend js "cvsctx.clearRect($sx,$sy,10,10);"
          }
          1 {
             lappend js "cvsctx.fillRect($sx,$sy,10,10);"
          }
        }
      }
    }
    jupyter::updatedisplay $jsid application/javascript [join $js \n];
    yield 1
  }
  yield 0
};
```
